## MACRO MOMENTUM



```python
yield2_tmp = web.DataReader('DGS2','fred', '1970-01-01') # Daily
yield2_tmp.index = yield2_tmp.index.set_names(["Date"])
```

```python
tickers = [ 'TLT']
prices_df = pd.DataFrame()

for tick in tickers:
    
    price = yf.download(tick,start='1990-01-01', progress = False, threads = False)
    price = price['Close']
    prices_df[tick] = price
    

comb_df = prices_df.merge(yield2_tmp, on="Date")    
```

```python
comb_df.loc[:,'ret'] = comb_df['TLT'].pct_change(1).shift(-1)
comb_df.loc[:,'2yr_yield_1Ychg'] = comb_df['DGS2'].pct_change(252)
comb_df = comb_df.dropna()
comb_df.loc[:,'SIGNAL'] = 0
comb_df.loc[comb_df['2yr_yield_1Ychg']<0,'SIGNAL'] = 1
comb_df.loc[:,'strat_ret'] = comb_df['SIGNAL'] * comb_df['ret']
```

```python
(comb_df[['ret','strat_ret']]+1).cumprod().plot();
```
